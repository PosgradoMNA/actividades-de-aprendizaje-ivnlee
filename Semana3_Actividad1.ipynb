### **Parte 1: Fundamentos de bases de datos**

**1. Fundamentos de bases de datos para ciencia de datos.** 
 
Los datos son posiblemente el recurso más importante en un proyecto de ciencia de datos. En este contexto, una base de datos se puede definir como un conjunto estructurado de datos almacenados en un medio electrónico (memoria o en la nube) de manera conveniente para que distintos usuarios puedan acceder a ellos.    
En general existen dos tipos de bases de datos dependiendo de la manera en la que la información está organizada: 
- *Base de datos relacional (Relational data base):* los datos o registros se organizan y almacenan en tablas que pueden conectarse entre sí.  

- *Base de datos no-relacional (NoSQL data base):* en este tipo de base de datos la información está conectada por categorías, utilizadas principalmente para datos no estructurados como video o imágenes. 

El acceso y manipulación de los datos se realiza mediante un lenguaje apropiado. Para bases de datos relacionales el más popular es SQL (Structured Query Language).  
 
 
**2. Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.** 
 
Un almacén de datos es un sistema centralizado que consolida grandes cantidades de información proveniente de diversas fuentes. Puede contener múltiples bases de datos, y habilita las tareas de gestión y análisis de datos para la toma de decisiones. La información se alimenta de manera regular y con el tiempo se almacenan datos históricos. Como punto importante, los datos se almacenan en formato tabular para favorecer el análisis. 


### **Parte 2: Selección y limpieza de los Datos en Python**

Importamos las librerías para trabajar con el conjunto de datos:

```python
import numpy as np
import pandas as pd
```

Definimos variable con liga al archivo .csv en github:

```python
input = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'
```

La descripcion de la base de datos la podemos ver aqui:
https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/dataset_info.txt

Creamos el conjunto de datos (data frame)

```python
df = pd.read_csv(input)
df 
```

Revisamos si existen valores faltantes en la base de datos:

```python
df.isnull().values.any()
```

```python
# Para detectar en que columnas faltan valores: 
df.isnull().any()
```

**A continuacion trataremos los valores faltantes**

- En caso de existir, eliminaremos las filas con todos los valores faltantes 

```python
# Creamos una copia del conjunto original para trabajar
df2 = df.copy()
# Metodo 'dropna' remueve valores faltantes
df2.dropna(how='all', inplace=True)
```

- Iniciaremos por la columna **Y:** representa la probabilidad real de que el cliente incumpla con el pago crediticio (1=Si, 0=No).  

```python
# verificamos valores únicos en la columna Y
df2['Y'].unique()
```

Optamos por eliminar las filas con valores nulos.

```python
# usamos parametro 'subset' para seleccionar las columnas deseadas
df2.dropna(inplace=True, subset=['Y'])
df2
```

```python
# verificamos
df2['Y'].isnull().values.any()
```

Podemos observar que se eliminaron 3 filas en total.

**Para el resto de las columnas X2 a X23 haremos imputación de valores**

- **X2:** género (1=masculino, 2=femenino). Verificamos que solo existan estos dos valores posibles e imputamos los faltantes con el valor de la moda.


```python
# verificamos valores únicos en la columna
df2['X2'].unique()
```

```python
# imputamos faltantes con la moda
x2mode = df2['X2'].mode()
df2['X2'].fillna(value=x2mode, inplace=True)

df2['X2'].unique()
```

- **X3:** nivel escolar (1=posgrado, 2=universidad, 3=preparatoria, 4=otros). similar a la columna X2 imputamos el valor de la moda a los faltantes. 

```python
# verificamos valores únicos en la columna
df2['X3'].unique()
```

Los valores 0, 5, y 6 no son válidos, por lo que los convertiremos a *NaN* y posteriormente imputaremos el valor de la moda usando solo valores permitidos (1, 2, 3 o 4).

```python
# convertimos a NaN los valores 
N = len(df2['X3'])
for i in range(N):
    if df2.iloc[i, 3] > 4:
        df2.iloc[i, 3] = np.nan 
    elif df2.iloc[i, 3] == 0:
        df2.iloc[i, 3] = np.nan

# verificamos
df2['X3'].unique()
```

```python
# imputamos faltantes con la moda
x3mode = df2['X3'].mode()
df2['X3'].fillna(value = x3mode, inplace=True)

df2['X3'].unique()
```

- **X4:** estado civil (1=casado, 2=soltero, 3=otro). Nuevamente imputamos el valor de la moda a los valores faltantes.

```python
# verificamos valores únicos en la columna
df2['X4'].unique() 
```

En este caso, el valor 0 no representa una categoria, por tanto lo que convertimos a nulo.

```python
# sustituimos los valores 0 por NaN
for i in range(N):
    if df2.iloc[i, 4] == 0:
        df2.iloc[i, 4] = np.nan 

df2['X4'].unique()
```

```python
# imputamos faltantes con la moda
x4mode = df2['X4'].mode()
df2['X4'].fillna(value = x4mode, inplace=True)

df2['X4'].unique()
```

- **X5:** edad del cliente (años)

```python
# Revisamos las estadisticas de las edades
df2['X5'].describe()
```

En esta caso decidimos imputar los valores faltantes con la mediana (no se observa gran diferencia con la media).

```python
# imputamos faltantes con la mediana
x5median = df2['X5'].median()
df2['X5'].fillna(value = x5median, inplace=True)

df2['X5'].isnull().values.any()
```

- **X6 - X11:** record de pago mensual de abril a septiembre (X6:septiembre, X7:agosto,...,X11:abril)

Valores: -1=pago a tiempo; 1=retraso 1 mes; 2=retraso 2 meses;...; 8=retraso 8 meses; 9=retraso 9 o mas meses.

```python
# revisamos columnas X6 a X11
x6_x11 = df[['X6', 'X7', 'X8', 'X9', 'X10', 'X11']]
x6_x11
```

Se observan valores de **-2.0** y **0** que no corresponden a ninguna categoría de estas columnas. Por lo tanto asignamos el valor *NaN* a todos los valores no permitidos. 

```python
# Iteramos fila por fila de la columna X6 a la X11
for j in range(6):
    for i in range(N):
        if df2.iloc[i, 6+j] < -1:
            df2.iloc[i, 6+j] = np.nan 
        elif df2.iloc[i, 6+j] == 0:
            df2.iloc[i, 6+j] = np.nan
        elif df2.iloc[i, 6+j] > 9:
            df2.iloc[i, 6+j] = np.nan
```

```python
# verificamos valores unicos 
print('X6: ', df2['X6'].unique())
print('X7: ', df2['X7'].unique())
print('X8: ', df2['X8'].unique())
print('X9: ', df2['X9'].unique())
print('X10: ', df2['X10'].unique())
print('X11: ', df2['X11'].unique())
```

Imputamos los nulos con el valor de la moda de cada columna.

```python
# calculamos la moda
x6mode = df2['X6'].mode()
x7mode = df2['X7'].mode()
x8mode = df2['X8'].mode()
x9mode = df2['X9'].mode()
x10mode = df2['X10'].mode()
x11mode = df2['X11'].mode()
```

```python
# imputamos valores
df2['X6'].fillna(x6mode, inplace=True)
df2['X7'].fillna(x7mode, inplace=True)
df2['X8'].fillna(x8mode, inplace=True)
df2['X9'].fillna(x9mode, inplace=True)
df2['X10'].fillna(x10mode, inplace=True)
df2['X11'].fillna(x11mode, inplace=True)

# verificamos
df2[['X6', 'X7', 'X8', 'X9', 'X10', 'X11']].isnull().values.any()
```

- **X12 - X17:** monto por pagar de abril a septiembre (X12:septiembre, X13:agosto,...,X17:abril)

```python
# revisamos estadisticas
amnt= df2[['X12', 'X13', 'X14', 'X15', 'X16', 'X17']]
amnt.describe() 
```

Observamos montos negativos, los cuales tendremos que asumir como error de captura y los convertimos a monto positivo.

```python
# usamos metodo abs para regresar solo valores positivos
for m in df2.iloc[:,12:18]:
    df2[m] = df2[m].abs()
 
# verificamos
amnt = df2[['X12', 'X13', 'X14', 'X15', 'X16', 'X17']]
amnt.describe()
```

Debido a que se observan valores extremos (*outliers*), usamos la mediana para los valores nulos.

```python
# calculamos la mediana
x12med = df2['X12'].median()
x13med = df2['X13'].median()
x14med = df2['X14'].median()
x15med = df2['X15'].median()
x16med = df2['X16'].median()
x17med = df2['X17'].median()
```

```python
# imputamos valores
df2['X12'].fillna(x12med, inplace=True)
df2['X13'].fillna(x13med, inplace=True)
df2['X14'].fillna(x14med, inplace=True)
df2['X15'].fillna(x15med, inplace=True)
df2['X16'].fillna(x16med, inplace=True)
df2['X17'].fillna(x17med, inplace=True)

# verificamos
df2[['X12', 'X13', 'X14', 'X15', 'X16', 'X17']].isnull().values.any()
```

- **X18 - X23:** monto de pago anterior de abril a septiembre (X18:septiembre, X19:agosto,...,X23:abril)

```python
# revisamos estadisticas
amnt2 = df2[['X18', 'X19', 'X20', 'X21', 'X22', 'X23']]
amnt2.describe() 
```

En este caso no hay valores negativos por lo que procedemos a imputar valores nulos con la mediana.

```python
# calculamos la mediana
x18med = df2['X18'].median()
x19med = df2['X19'].median()
x20med = df2['X20'].median()
x21med = df2['X21'].median()
x22med = df2['X22'].median()
x23med = df2['X23'].median()
```

```python
# imputamos valores
df2['X18'].fillna(x18med, inplace=True)
df2['X19'].fillna(x19med, inplace=True)
df2['X20'].fillna(x20med, inplace=True)
df2['X21'].fillna(x21med, inplace=True)
df2['X22'].fillna(x22med, inplace=True)
df2['X23'].fillna(x23med, inplace=True)

# verificamos
df2[['X18', 'X19', 'X20', 'X21', 'X22', 'X23']].isnull().values.any()
```

Verificamos nuevamente que no existan columnas con valores nulos en el data frame:

```python
df2.isnull().any() 
```

**Finalmente tenemos un *data frame* sin valores nulos:**

```python
# data frame final:
print(df2.shape)

df2
```
